```python
# Table 1: Nutrient (nitrogen and phosphorus) retention and excretion rates (as a percentage of the feed consumed). Modified from Piedrahita 2003.
table1_data = pd.read_csv('./Pollution_rejectants/Data/Table1_N_P_retained_excreted.csv', header=0, decimal = ',')
table1_data.columns = ['Chemicals', 'Exiting route', '% of ingested']
table1_data.Chemicals = table1_data.Chemicals.apply(lambda x: 'Nitrogen' if x == 'Azote'
                                          else 'Phosphorus')
table1_data['Exiting route'] = table1_data['Exiting route'].apply(lambda x: 'Retained as flesh' if x == 'retained_as_flesh'
                                          else 'Excreted as Feces' if x == 'Undigested_particulate_matter'
                                          else 'Excreted as dissolved form')


sns.set(rc={"figure.figsize":(15, 6)}) #width=*15, height=6
table1 = sns.barplot(table1_data, x = 'Exiting route', y = '% of ingested', hue='Chemicals').set(title='Exiting route of Nitrogen and Phosphorus after ingestion by Salmo Salar (%)')

```

Propose une liste des manques dans la surveillance et la gestion environnementale des fermes marines de saumons

The Chilean law considers projects entering the EIA for evaluation as Estudio de Im-
pacto Ambiental (Environmental Impact Study, EIS) and Declaración de Impacto Ambiental
(Environmental Impact Declaration, EID)

*An EIS is a document where all the characteristics of the project—to be executed—are described in detail. The report must contain all background information on prediction, identification and interpretation of the environmental impacts, and must describe how significant adverse effects are prevented or minimized. 
An EID, on the other hand, is a project report, under oath of the owner, where its content allows for an environmental impact evaluation in accordance with the environmental law*

Les principaux pathogènes visés sont caligidosis (*Caligus rogercresseyi*), sea lice and *Piscirickettsia salmonis*
Pour le moment, ils utilisent : immersion treatments with organophosphates and pyrethroids, and in recent years it has been complemented with therapies based to chitin synthesis inhibitors, hydrogen peroxide and non-pharmacological treatments. Les parasites commencent à développer des résistances à ces traitements.

### Sea farming

L'indice de consommation = gr atb utilisés en 2021/biomasse produite (récupérées + morte) en tonnes. La moyenne, toutes companies confondues est de 394.1 gr atb_tonnes de biomasse produite.

Index, indice_atb: gr atb/tonnes(biomasse produite + biomasse morte) différent de plus haut qui était un % en tonne/tonne. Du coup, j'ai refait une colonne tonne/tonne.
biomasse produite: tonnes selon traçabilité des entreprises. 
biomasse morte: calculée à partir de la somme des poissons morts dans chaque cage et des poids moyens (selon information declarées à la SIFA).

Table 5: Quantité d\antibiotiques, de production, de biomasse morte et indice de consommation d'antibotique en 2021 au Chili.

```python


# Table 5 : Quantité d\antibiotiques, de production, de biomasse morte et indice de consommation d'antibotique en 2021 au Chili.
table5_data = pd.read_csv('./Pollution_rejectants/Data/Table5_Chili_2021_companies_prod_atb.csv', header=0, decimal = ',')

table5_data['biomasse_vivante'] = table5_data.production_ton - table5_data.biomasse_morte_ton
table5_data['ica_en%'] = round(table5_data.ton_principe_actif/table5_data.production_ton *100, 3)
display(table5_data)

sns.set(rc={"figure.figsize":(30, 10)}) #width=*15, height=6
table5 = sns.barplot(table5_data, x = 'companie', y = 'ton_principe_actif', hue='production_ton', orient='v').set(title='Table 5 : Quantité d\'antibiotiques, de production, de biomasse morte et indice de consommation d\'antibotique en 2021 au Chili.')
```
